```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load audio
debussy_file = "audio/debussy.wav"
redhot_file = "audio/redhot.wav"
duke_file = "audio/duke.wav"

debussy, sr = librosa.load(debussy_file)
redhot, _ = librosa.load(redhot_file)
duke, _ = librosa.load(duke_file)
```

```python
FRAME_LENGTH = 1024
HOP_LENGHT = 512

zcr_debussy = librosa.feature.zero_crossing_rate(debussy,frame_length=FRAME_LENGTH, hop_length=HOP_LENGHT)[0]
zcr_redhot = librosa.feature.zero_crossing_rate(redhot,frame_length=FRAME_LENGTH, hop_length=HOP_LENGHT)[0]
zcr_duke = librosa.feature.zero_crossing_rate(duke,frame_length=FRAME_LENGTH, hop_length=HOP_LENGHT)[0]
```

```python
# Vizualice all the ZCR for each song

frames = range(0,len(zcr_debussy))
t = librosa.frames_to_time(frames,hop_length=HOP_LENGHT)


plt.figure(figsize=(15,17))

plt.plot(t,zcr_debussy, color="r")
plt.plot(t,zcr_redhot, color="y")
plt.plot(t,zcr_duke, color="g")
plt.ylim((0,1))

plt.show()
```
